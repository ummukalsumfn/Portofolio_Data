Check Outliers

```python
# Before (Boxplot)
airbnb.boxplot(column='price')
```

```python
airbnb['price'].describe() #Statistik Deskriptif
```

Batas Atas= Q3 + 1.5 X (Q3-Q1), Batas Bawah= Q1-1.5 (Q3-Q1)

```python
q1=airbnb['price'].quantile(0.25)
q3=airbnb['price'].quantile(0.75)
upper_limit=q3+1.5*(q3-q1)
lower_limit=q1-1.5*(q3-q1)
```

```python
airbnb=airbnb[airbnb['price'] > lower_limit]
airbnb=airbnb[airbnb['price'] < upper_limit]
airbnb['price'].describe()
```

```python
#After
airbnb.boxplot(column='price')
```

```python
airbnb.info()
```

Jumlah data berubah dari 48874 menjadi 45897. Berarti jumlah outlier ada sekitar 3000an. Jika dibandingkan 3000/48000 = 0.0625 (tidak sampai 10%nya) jumlah outlier tidak signifikan maka bisa dihapus.
